```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
import time
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException, TimeoutException, WebDriverException
import logging
from seleniumbase import Driver
import pandas as pd
```

```python
team_links = {
    "Islamabad United": "https://www.espncricinfo.com/records/trophy/averages-batting/pakistan-super-league-205?team=5795",
    "Karachi Kings": "https://www.espncricinfo.com/records/trophy/averages-batting/pakistan-super-league-205?team=5793",
    "Lahore Qalandars": "https://www.espncricinfo.com/records/trophy/averages-batting/pakistan-super-league-205?team=5799",
    "Multan Sultans": "https://www.espncricinfo.com/records/trophy/averages-batting/pakistan-super-league-205?team=6413",
    "Peshawar Zalmi": "https://www.espncricinfo.com/records/trophy/averages-batting/pakistan-super-league-205?team=5791",
    "Quetta Gladiators": "https://www.espncricinfo.com/records/trophy/averages-batting/pakistan-super-league-205?team=5797"
}
```

```python
all_team_data = []

for team, link in team_links.items():
    driver.get(link)
    table = driver.find_element(By.XPATH, '//table[1]')
    rows = table.find_elements(By.TAG_NAME, 'tr')
    team_data = []
    
    for row in rows[1:]:
        cells = row.find_elements(By.TAG_NAME, "td")
        if len(cells) > 0:
            row_data = {
                "Team": team,
                "Player": cells[0].text,
                "Span": cells[1].text,
                "Mat": cells[2].text,
                "Inns": cells[3].text,
                "NO": cells[4].text,
                "Runs": cells[5].text,
                "HS": cells[6].text,
                "Ave": cells[7].text,
                "SR": cells[8].text,
                "100": cells[9].text,
                "50": cells[10].text,
                "0": cells[11].text
            }
            team_data.append(row_data)
    
    all_team_data.extend(team_data)
```

```python
batting_df = pd.DataFrame(all_team_data)
batting_df
```

```python
batting_df.to_csv('PSL_Batting_Dataset.csv')
```
